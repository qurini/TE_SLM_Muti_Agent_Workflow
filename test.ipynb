```python
import json

def read_jsonl_basic(filepath):
    data = []
    with open(filepath, 'r') as f:
        for line in f:
            try:
                data.append(json.loads(line.strip()))
            except json.JSONDecodeError as e:
                print(f"Error decoding JSON on line: {line.strip()} - {e}")
    return data

# Example usage
# Assuming 'my_data.jsonl' exists with valid JSONL content
# my_data.jsonl:
# {"name": "Alice", "age": 30}
# {"name": "Bob", "age": 25}
# parsed_data = read_jsonl_basic('my_data.jsonl')
# for item in parsed_data:
#     print(item)
```
